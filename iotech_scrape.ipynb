```python
import requests 
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
from time import sleep
```

```python
url='https://bbs.io-tech.fi/'
page = requests.get(url)
soup = bs(page.text, 'html.parser')
links = soup.find_all(href=True)
# Select which titles will be used from https://bbs.io-tech.fi/
titles=[link['href'] for link in links[44:49]]
```

```python
def get_titles(category,cat_num):
    category_titles=[]
    page_num=1
    while True:
        if page_num==1:
            title_url='https://bbs.io-tech.fi/'+category
        else:
            title_url='https://bbs.io-tech.fi/'+category+'page'+'-'+str(page_num)  
        sub_page = requests.get(title_url)
        if sub_page.status_code!=200 or page_num>=2:
            break
        soup = bs(sub_page.text, 'html.parser')
        titles = soup.find_all('div', {'class': 'structItem-title'})
        stats = [
        (i.contents[0], i.contents[1].text) for i in titles]
        for i in stats:
            category_titles.append(i[1])
        page_num+=1
    frame=pd.DataFrame(category_titles, columns=['Titles'])
    frame['Category']=cat_num
    return frame
```

```python
full_data2=pd.DataFrame()
for count,title in enumerate(titles):
    full_data2=pd.concat([full_data2,get_titles(title,count)])
```
